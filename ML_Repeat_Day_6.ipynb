## **Procurement Strategy Predictor**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score , classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import pickle
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shahriarkabir/procurement-strategy-dataset-for-kraljic-matrix")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(path + "/realistic_kraljic_dataset.csv")
df.head(5)
```

```python
df["Kraljic_Category"].value_counts()
```

```python
df["Supplier_Region"].value_counts()
```

```python
df.info()
```

```python
df["Single_Source_Risk"] = df["Single_Source_Risk"].map({"No": 0, "Yes": 1})
```

```python
df.head(10)
```

```python
x = df.drop(["Product_ID","Supplier_Region","Product_Name", "Kraljic_Category"], axis=1)
y = df["Kraljic_Category"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
# Call the model

model_knn = KNeighborsClassifier()
model_nb = GaussianNB()
model_lr = LogisticRegression()
model_svm = SVC()
```

```python
model_knn.fit(x_train,y_train)
model_nb.fit(x_train,y_train)
model_lr.fit(x_train,y_train)
model_svm.fit(x_train,y_train)
```

```python
y_pred_knn = model_knn.predict(x_test)
y_pred_nb = model_nb.predict(x_test)
y_pred_lr = model_lr.predict(x_test)
y_pred_svm = model_svm.predict(x_test)
```

```python
print("KNN Report :\n",classification_report(y_test,y_pred_knn))
print("Navie Report :\n",classification_report(y_test,y_pred_nb))
print("Logistic Report :\n",classification_report(y_test,y_pred_lr))
print("SVM Report :\n",classification_report(y_test,y_pred_svm))
```

```python
# test the model
scores = {
    "KNN": model_knn.score(x_test, y_test),
    "Logistic Regression": model_lr.score(x_test, y_test),
    "SVM": model_svm.score(x_test, y_test)
}

best_model = max(scores, key=scores.get)
print(f"\n✅ Best Model is: {best_model} with Accuracy = {scores[best_model]:.2f}")
```

```python
# Create a PKL file
with open('Procurement.pkl', 'wb') as f:
    pickle.dump(model_lr, f)
```
